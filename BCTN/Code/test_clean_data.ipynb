```python
import re

def convert_currency_to_number(currency_string):
    if 'Tỷ' in currency_string:
        number = int(float(re.search(r'[\d.]+', currency_string).group()) * 10**9)
    elif 'Triệu' in currency_string:
        number = int(float(re.search(r'[\d.]+', currency_string).group()) * 10**6)
    else:
        number = int(float(re.search(r'[\d.]+', currency_string).group()))
    return number

# Assuming data is a Pandas DataFrame
data["Đơn giá"] = data["Đơn giá"].apply(lambda x: convert_currency_to_number(str(x)) if pd.notnull(x) else x)
```
